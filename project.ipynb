```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
%matplotlib inline
```

```python
ls = os.listdir('stockData')
lsLen = len(ls)
for x in range(lsLen):
    ls[x] = ls[x].split('.')[0].replace('-','_')
```

```python
def fileExists(fileName):
    lsImg = os.listdir('images')
    lsLenImg = len(lsImg)
    exists = False
    for x in range(lsLenImg):
        lsImg[x] = lsImg[x].split('.')[0]+'.png'
        if lsImg[x].find(fileName)>-1:
            exists = True
    return exists
```

```python
def csvToImage(fileName):
    if(not len(fileName)<=0):        
        df = pd.read_csv('stockData/'+fileName+'.csv')
        high = df.High
        low = df.Low
        date = df.Date #x-axis/independent variable
        mid = ((high+low)/2.0)

        fig = plt.figure(figsize=(11,8), dpi=1000)
        fig, axes = plt.subplots(figsize=(11,8))

        plt.ylabel('Mid EOD Value')
        plt.xlabel('Date')
        plt.title('Mid EOD vs Date for: '+fileName)

        axes.plot(date,mid,'b')
        plt.show()
        
        # remove if image exists
        if fileExists(fileName):
            os.remove('images/mid_'+fileName+'.png')

        # then recreate
        fig.savefig('images/mid_'+fileName+'.png')
```

```python
for x in range(lsLen):
    csvToImage(ls[x])
```
